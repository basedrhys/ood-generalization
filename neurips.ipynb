```python
import pandas as pd
pd.options.display.max_rows = 100
pd.options.display.max_columns = 30

envs = ["MIMIC", "CXP", "NIH", "PAD"]

df = pd.read_csv("wandb_export_2022-09-21T09_48_45.621-04_00.csv")
df["balance_resample"] = df["balance_resample"].fillna("none")
df
```

```python
df_pn = df[df["binary_label"] == "All"]
df_pn
```

```python
grouped.loc[("CXP", "CXP")].sort_index(ascending=False).diff()
```

```python
f1 = lambda x: f"test.{x}_f1"
f2 = lambda x: f"test.{x}_roc"
f3 = lambda x: f"test.{x}_acc"

metric_cols = [f(x) for x in envs for f in (f1, f2, f3)]

grouped = df_pn.groupby(["test_env", "train_envs", "balance_resample"]).agg("mean")[metric_cols]
grouped
```

```python
diffs = grouped.sort_index(ascending=[True, True, False]).diff()
diffs
```

```python
idx = pd.IndexSlice
diffs.loc[idx[:, :, "label,under"], :]
```

```python
def display_balanced_results(env):
    return grouped.loc[env][[f"test.{env}_f1", f"test.{env}_roc", f"test.{env}_acc"]]

display_balanced_results("MIMIC")
```
